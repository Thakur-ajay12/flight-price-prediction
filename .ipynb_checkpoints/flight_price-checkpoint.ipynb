```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train_data=pd.read_excel(r"Data_Train.xlsx")
```

```python
pd.set_option('display.max_columns',None)
```

```python
train_data.head()
```

```python
train_data.info()
```

```python
train_data.shape
```

```python
# how many unique value are there
train_data["Duration"].value_counts()
```

```python
# drop none value
train_data.dropna(inplace=True)
```

```python
train_data.shape
```

```python
train_data.isnull().sum()
```

```python
#   format="%d/%m/%Y" this is not much imp code will work without this
train_data["Journey_day"] = pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y" ).dt.day
```

```python
train_data['journey_month'] = pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
# In-place operation is an operation that changes directly the content of a given linear algebra, vector, matrices (Tensor)without making a copy

train_data.drop(["Date_of_Journey"],axis=1, inplace=True)
```

```python
train_data["dep_hour"] = pd.to_datetime(train_data.Dep_Time).dt.hour
```

```python
train_data["dep_time"] = pd.to_datetime(train_data.Dep_Time).dt.minute
```

```python
train_data.drop(["Dep_Time"], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
train_data.head()
```

```python
# Arrival time is when the plane pulls up to the gate.
# Similar to Date_of_Journey we can extract values from Arrival_Time

# Extracting Hours
train_data["Arrival_hour"] = pd.to_datetime(train_data.Arrival_Time).dt.hour

# Extracting Minutes
train_data["Arrival_min"] = pd.to_datetime(train_data.Arrival_Time).dt.minute

# Now we can drop Arrival_Time as it is of no use
train_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
# Time taken by plane to reach destination is called Duration
# It is the differnce betwwen Departure Time and Arrival time


# Assigning and converting Duration column into list
duration = list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

```

```python
#create two list var 
duration_hours = []
duration_mins = []
```

```python
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1] ))   # Extracts only minutes from duration
```

```python
# Adding duration_hours and duration_mins list to train_data dataframe

train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
train_data.head()
```

```python
train_data.drop(["Duration"],axis=1,inplace=True)
```

```python
train_data["Airline"].value_counts()
```

```python
# From graph we can see that Jet Airways Business have the highest Price.
# Apart from the first Airline almost all are having similar median

# Airline vs Price
sns.catplot(y = "Price", x = "Airline", data = train_data.sort_values("Price", ascending = False), kind="boxen", height = 6, aspect = 3)
plt.show()
```

```python
# As Airline is Nominal Categorical data we will perform OneHotEncoding

Airline = train_data[["Airline"]]

Airline = pd.get_dummies(Airline, drop_first= True)

Airline.head()
```

```python
train_data["Source"].value_counts()
```

```python
# Source vs Price
# in airline we use catplot but here we use boxplot just to explore

sns.boxplot(y = "Price", x = "Source", data = train_data.sort_values("Price", ascending = False))
plt.show()
```

```python
Source = train_data[["Source"]]
Source = pd.get_dummies(Source, drop_first=True)
Source.head()
```

```python
Destination = train_data["Destination"]
Destination = pd.get_dummies(Destination, drop_first=True)
Destination.head()
```

```python
train_data["Total_Stops"].value_counts()
```

```python
# As this is case of Ordinal Categorical type we perform LabelEncoder
# Here Values are assigned with corresponding keys

train_data.replace({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4},inplace=True)
```

```python
train_data.head()
```

```python
# Concatenate dataframe --> train_data + Airline + Source + Destination
# now final data is in new var data_train

data_train = pd.concat([train_data,Airline,Source,Destination],axis=1)
```

```python
data_train.head()
```

```python
data_train.drop(["Airline", "Source", "Destination"],axis=1, inplace=True)
data_train.head()
```
